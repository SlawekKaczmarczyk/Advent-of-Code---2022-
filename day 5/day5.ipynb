```python
# input1 is a multiline string

stacklines, movements = input1.split('\n\n')
stacklines = stacklines.splitlines()
movements = movements.splitlines()

# adapt to larger warehouses (warehice?) by calculating the number of stacks
num_of_stacks = int(max(stacklines[-1].split()))

stacksh = []    # gravity-defying horizontal "stacks"
stacksv = []    # proper vertical stacks

for line in stacklines:
    l = []
    for i in range(2, num_of_stacks * 4, 4):
        l.append(line[i - 1])
    stacksh.append(l)

for i in range(len(stacksh[0])):
    l = []
    for j in range(len(stacksh) - 1):
        if stacksh[j][i] != ' ':
            l.append(stacksh[j][i])
    stacksv.append(l)

for move in movements:
    move_count = int(move.split()[1])
    stack_from = int(move.split()[3]) - 1
    stack_to = int(move.split()[5]) - 1
    for i in range(move_count):
        stacksv[stack_to].insert(i, stacksv[stack_from][0])
        del stacksv[stack_from][0]

res = ''

for stack in stacksv:
    res += stack[0]

print(res)
```
